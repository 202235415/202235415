```python
data2 = pd.read_excel('C:/Users/test4/Desktop/대학 강의/2-2/데이터처리프로그래밍/dataset/04_data1.xlsx', index_col='area')
data2.head()
```

```python
data2020_good = data2['2020_good']
data2019_good = data2['2019_good']
```

```python
plt.figure(figsize=(20,4))
plt.plot(data2020_good, color='y', marker='o')
plt.plot(data2019_good, color='r', marker='+')
plt.xlabel('Area')
plt.ylabel('Micrometer')
plt.title('2019 & 2020 FIne Dust(pm10) Good Line Graph')
plt.grid()
plt.show()
```

```python
data2015_bad = data2['2015_bad']
data2020_bad = data2['2020_bad']
plt.figure(figsize=(20,4))
plt.bar(data2020_bad.index, data2020_bad, color='b', alpha=0.5)
plt.bar(data2015_bad.index, data2015_bad, color='g', alpha=0.5)
plt.xlabel('Area')
plt.ylabel('Micrometer')
plt.title('2015 & 2020 FIne Dust(pm10) Bad Bar Graph')
plt.grid()
plt.show()
```

```python
import numpy as np
index = np.arange(7)

plt.figure(figsize=(20,4))

data3 = data2.loc['Seoul':'Ulsan', '2015_good':'2020_good']
for year in range(2015,2021):
    year = str(year)+'_good'
    chartdata = data3[year]
    plt.bar(index, chartdata, width=0.15, label=year)
    index = index + 0.15
    
plt.xlabel('Area')
plt.ylabel('Micrometer')
plt.xticks(index-0.53, ['Seoul', 'Busan', 'Dargu', 'Incheon', 'Gwangju', 'Darjeon', 'Ulsan'])
plt.title('2015~2020 Fine Dust(pm10) Good Group Bar Graph')
plt.legend()
plt.show()
```

```python
index = np.arange(7)

plt.figure(figsize=(20,4))

data4 = data2.loc['Seoul':'Ulsan', '2017_common':'2020_common']
for year in range(2017,2021):
    year = str(year)+'_common'
    chartdata = data4[year]
    plt.barh(index, chartdata, width=0.15, label=year)
    

plt.xlim(120,310)    
plt.xlabel('Area')
plt.ylabel('Micrometer')
plt.yticks(index, ['Seoul', 'Busan', 'Dargu', 'Incheon', 'Gwangju', 'Darjeon', 'Ulsan'])
plt.title('2017~2020 Fine Dust(pm10) Common Group Barh Graph')
plt.legend()
plt.show()
```

```python
iris = pd.read_csv('C:/Users/test4/Desktop/대학 강의/2-2/데이터처리프로그래밍/dataset/04_data2.csv')
iris.head()
```

```python
PetalLength = iris['PetalLength']
PetalWidth = iris['PetalWidth']

plt.figure(figsize=(10,4))
plt.scatter(PetalLength, PetalWidth)

plt.xlabel=('Peter Length')
plt.ylabel=('Peter Width')
plt.title('Iris Petal Scatter Graph')
plt.grid()
plt.show()
```

```python
setosa = iris.loc[iris.Species=='setosa',['Species','SepalWidth']]
versicolor = iris.loc[iris.Species=='versicolor',['Species','SepalWidth']]

plt.figure(figsize=(10,6))
plt.hist(setosa['SepalWidth'], bins=5, alpha=0.5, label='setosa', color='r')
plt.hist(versicolor['SepalWidth'], bins=5, alpha=0.5, label='versicolor', color='b')

plt.xlabel=('Sepal')
plt.ylabel=('Width')
plt.title('Iris Sepal Group Histogram')
plt.legend()
plt.grid()
plt.show()
```
