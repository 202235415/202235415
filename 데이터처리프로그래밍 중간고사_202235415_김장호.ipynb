```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import time
import seaborn as sns
```

```python
df = pd.read_csv('C:/Users/test4/Desktop/대학 강의/2-2/데이터처리프로그래밍/중간고사/test1026.csv')
df
```

```python
df['pickup_datetime'] = pd.to_datetime(df['pickup_datetime'])
df['dropoff_datetime'] = pd.to_datetime(df['dropoff_datetime'])
```

```python
df['weekday'] = df.pickup_datetime.dt.day_name()
df['month'] = df.pickup_datetime.dt.month
df['weekday_num'] = df.pickup_datetime.dt.weekday
df['pickup_hour'] = df.pickup_datetime.dt.hour
```

```python
passenger_count_sum = df.groupby(['vendor_id', 'weekday_num'])['passenger_count'].sum().reset_index()

plt.figure(figsize=(20, 6))
sns.barplot(x='weekday_num', y='passenger_count', hue='vendor_id', data=passenger_count_sum)

plt.xlabel('passenger_count')
plt.ylabel('count')
plt.title('Passenger per Hour using Count Plot Graph')
plt.legend()
plt.show()
```

```python
newdf = df.loc[:, ['pickup_hour', 'passenger_count', 'vendor_id', 'month']]
```

```python
plt.figure(figsize=(13,10))
plt.title('Heat Map Graph')
correlation_newdf = newdf.corr()
mat = np.triu(correlation_newdf)
sns.heatmap(correlation_newdf, annot=True, cmap='YlGnBu', mask=mat)
```

```python
hist1=df.loc[df['month'] == 1,['passenger_count','weekday_num']]
hist2=df.loc[df['month'] == 5,['passenger_count','weekday_num']]

plt.figure(figsize=(10,6))
plt.hist(hist1['weekday_num'], bins=5, alpha=0.4, label='JAN')
plt.hist(hist2['weekday_num'], bins=5, alpha=0.4, label='FEB')

plt.xlabel('weekday')
plt.ylabel('Passenger Count')
plt.title('Passenger Count Per WEEKDAY')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(hist1['weekday_num'], bins=5, alpha=0.4, label='JAN',kde=True)
sns.histplot(hist2['weekday_num'], bins=5, alpha=0.4, label='FEB',kde=True,color='r')

plt.xlabel('weekday')
plt.ylabel('Passenger Count')
plt.title('Passenger Count Per WEEKDAY')
plt.legend()
plt.grid()
plt.show()
```
