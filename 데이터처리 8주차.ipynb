```python
# 데이터 시각화
# 개봉년도별 드라마수
import numpy as np
year_Count=newdrama.groupby('개봉년도').size()
plt.figure(figsize=(16,10))
plt.plot(year_Count,color='r',marker='o')
plt.bar(year_Count.index, year_Count, color='g')
plt.xticks(np.arange(2003,2022))
plt.xlabel('Year of release')
plt.ylabel('Count')
plt.grid()
plt.show()
```

```python
# 개봉년도별 평점의 평균에 대한 시각화
import matplotlib.pyplot as plt
import numpy as np

data=drama.groupby('Year of release')['Rating'].mean()
plt.figure(figsize=(16,10))
plt.plot(data,color='r',marker='o')
plt.xticks(np.arange(2003,2022))
plt.xlabel('Year of release')
plt.ylabel('Rating Average')
plt.grid()
plt.show()
```

```python
#장르별 빈도수 구하기
from collections import Counter
newdrama['장르'] = newdrama['장르'].str.strip()
genre_list = list()
for genres in newdrama['장르'].to_list():
    genres = genres.split(",  ")
    for gen in genres:
        genre_list.append(gen)       
genre_df = pd.DataFrame.from_dict(Counter(genre_list), orient = 'index').rename(columns = {0:'Count'})
genre_df.sort_values('Count',ascending = False, inplace = True)
genre_df.head()
```

```python
plt.figure(figsize=(25,10))
plt.plot(genre_df,color='r',marker='o')
plt.bar(genre_df.index,genre_df.Count)
plt.title('Genre / Drame Count')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.show()
```

```python
actor_list = []
for actors in drama['Cast']:
    actors = actors.split(", ")
    actor_list.extend(actors)

actor_count = dict(Counter(actor_list))

# 카운트된 결과를 데이터프레임으로 변환하여 정렬
actor_count_df = pd.DataFrame.from_dict(actor_count, orient='index', columns=['출연수']).reset_index()
actor_count_df.columns = ['출연배우', '출연수']
actor_count_df.sort_values('출연수', ascending=False, inplace=True)
pd.DataFrame(actor_count_df)
```
