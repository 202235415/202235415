```python
import pandas as pd
data = pd.read_excel('C:/Users/test4/Desktop/대학 강의/2-2/데이터처리프로그래밍/dataset/04_data1.xlsx', index_col='area')
data.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt
index =np.arange(7)

plt.figure(figsize=(10,4))

data2 = data.loc['Seoul':'Ulsan', '2017_common':'2020_common']
for year in range(2017,2021):
    year = str(year) + '_common'
    chartdata = data2[year]
    plt.barh(index, chartdata, label=year)
    
plt.xlim(120,310)
plt.ylabel('Area')
plt.xlabel('Micrometer')
plt.yticks(index, ['Seoul', 'Busan', 'Daegu', 'Incheon', 'Gwangju', 'Daejeon', 'Ulsan'])
plt.title('2017~2020 Fine Dust(pm10) Common Group Barh Graph')
plt.legend()
plt.show()
```

```python
titanic = pd.read_csv("C:/Users/test4/Desktop/대학 강의/2-2/데이터처리프로그래밍/dataset/titanic.csv")
titanic.head()
```

```python
titanic.isnull().sum()
```

```python
titanic['Embarked'].fillna(titanic['Embarked'].mode()[0], inplace =True)
titanic.drop('Cabin',axis=1,inplace=True)
avg=titanic['Age'].mean()
titanic['Age'].fillna(avg,inplace=True)
titanic.isnull().sum()
```

```python
Fare = titanic['Fare']
Pclass = titanic['Pclass']

plt.figure(figsize=(10,4))
plt.scatter(Pclass, Fare, color = 'r')

plt.xlabel('Pclass')
plt.ylabel('Fare')
plt.title('Pclass & Fare Scatter Graph')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist(titanic['Fare'], bins=8, label='Fare', color='k')

plt.xlabel('Fare')
plt.ylabel('Passenger')
plt.title('Fare Histogram')
plt.legend()
plt.grid()
plt.show()
```

```python
newdf = titanic.copy().head(100)
newdf
```

```python
plt.figure(figsize=(10,6))
plt.hist(newdf['Fare'], bins=8, label='Fare', color='m')

plt.xlabel('Fare')
plt.ylabel('Passsenger')
plt.title('100 Passenger Fare Histogram')
plt.legend()
plt.grid()
plt.show()
```

```python
age = newdf.loc[newdf.Survived==1,['Survived','Age']]
fare = newdf.loc[newdf.Survived==1,['Survived','Fare']]

plt.figure(figsize=(10,6))
plt.hist(age['Age'], bins=50, alpha=0.5, label='Age')
plt.hist(fare['Fare'], bins=50, alpha=0.5, label='Fare')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Survived Age & Fare Group Histogram')
plt.legend()
plt.grid()
plt.show()
```

```python
age2 = newdf['Age']
fare2 = newdf['Fare']

plt.figure(figsize=(8,5))
plt.boxplot([age2], labels=['Age'])

plt.ylabel('Value')
plt.title('Age Box Plot')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.boxplot([fare2], labels=['Fare'])

plt.ylabel('Value')
plt.title('Fare Box Plot')
plt.show()
```
